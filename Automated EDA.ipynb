# to know and handle path of the file 


```python
def file_path(path):
    file_name,file_extension=os.path.splitext(path)
    if file_extension=='.csv':
        df=pd.read_csv(path)
    elif file_extension=='.xlsx':
        df=pd.read_excel(path)
    else:
        raise ValueError("Unsupported file format. Please provide a CSV or XLS/XLSX file.")
    return df
data=file_path("/home/mohamed/Downloads/archive (1)/International_Report_Departures.csv")
data
# /home/mohamed/Downloads/archive (1)/International_Report_Departures.csv
#/home/mohamed/Downloads/world_bank_development_indicators.csv"
```

# preprocessing the data

```python
def preprocess(df):
    # Imputing missing values using median
    numerical_columns = df.select_dtypes(include=['number']).columns
    if not numerical_columns.empty:
        # If there are numerical columns
        df.fillna(data.median(numeric_only=True), inplace=True)
        
        # Normalize numerical columns using Min-Max Scaling (noramlization)
        for column in numerical_columns:
            df[column] = (df[column] - df[column].min()) / (df[column].max() - df[column].min())    

    return df.head()

data_processed = preprocess(data)
data_processed
```

# visualize data

```python
import ipywidgets as widgets
from IPython.display import display
```

```python
# Function to generate visualizations based on column types
def generate_visualizations(column):
    if data[column].dtype == 'object':
        plt.figure(figsize=(10, 6))
        sns.countplot(data=data, x=column)
        plt.xticks(rotation=45)
        plt.title(f'Count plot of {column}')
        plt.show()
    elif data[column].dtype == 'float64' or data[column].dtype == 'int64':
        plt.figure(figsize=(10, 6))
        sns.histplot(data=data, x=column, bins=10)
        plt.title(f'Histogram of {column}')
        plt.show()

# Dropdown widget for selecting columns
column_selector = widgets.Dropdown(
    options=data.columns,
    description='Select a column:',
    disabled=False,
)

# Button to trigger visualization
generate_button = widgets.Button(description="Generate Visualization")

# Output area for displaying the visualization
output = widgets.Output()

# to generate visualization on button click
def on_generate_button_clicked(b):
    output.clear_output()
    with output:
        generate_visualizations(column_selector.value)

generate_button.on_click(on_generate_button_clicked)

# Display widgets and output
display(column_selector, generate_button, output)
```
